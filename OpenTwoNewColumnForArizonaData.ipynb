```python
# Get a list of all the uploaded csv files in the current directory of the colab notebook
files = [f for f in os.listdir() if f.endswith('.csv')]
```

```python
# Process each file
for file_name in files:
    # Split the filename to extract city and zipcode
    parts = file_name.split('_')
    city = parts[0]
    zipcode = parts[1]

    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_name)

    # Add the new 'zipcode' and 'city' columns to the DataFrame
    df['zipcode'] = zipcode
    df['city'] = city

    # Save the modified DataFrame back to the same CSV file
    df.to_csv(file_name, index=False)
```

```python
import os
import zipfile
from google.colab import files
```

```python
directory = '/content/'
```

```python
zip_filename = "Arizona_processed_files.zip"
```

```python
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    for foldername, subfolders, filenames in os.walk(directory):
        for filename in filenames:
            # Ensure you're not trying to zip the zip file itself (if it already exists)
            if filename != zip_filename:
                zipf.write(os.path.join(foldername, filename), os.path.relpath(os.path.join(foldername, filename), directory))
```

```python
files.download(zip_filename)
```

```python
csv_files = [f for f in os.listdir() if f.endswith('.csv')]
```

```python
# Initialize an empty list to hold DataFrames
dfs = []
```

```python
# Read each CSV into a DataFrame and append to the list
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    dfs.append(df)
```

```python
# Merge all DataFrames into one
merged_df = pd.concat(dfs, ignore_index=True)
```

```python
# Save the merged DataFrame into a new CSV file
merged_df.to_csv("Modified_Arizona_Data.csv", index=False)
```

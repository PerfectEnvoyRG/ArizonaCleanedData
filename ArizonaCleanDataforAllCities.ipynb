Get a list of all files in the '/content/' directory

```python
import os
import pandas as pd
```

```python
files = os.listdir('/content/')
```

Columns to keep

```python
cols_to_keep = ['Program Name', 'Implementing Sector', 'Category' , 'State' , 'Incentive Type' , 'Web Site' , 'Applicable Sectors' , 'Eligible Renewable/Other Technologies' , 'Administrator' , 'Eligible Storage Technologies']

def combine_columns(row):
    combined_data = []
    for col in cols_to_combine:
        value = row[col]
        combined_data.append(f"{col}: {value if pd.notna(value) else 'None'}")
    return '; '.join(combined_data)

for file in files:
    file_path = os.path.join('/content/', file)

    # Check if the path is a file (and not a directory)
    if os.path.isfile(file_path):
        try:
            df = pd.read_csv(file_path)

            # Columns to combine
            cols_to_combine = [col for col in df.columns if col not in cols_to_keep]

            df['Additional Details'] = df.apply(combine_columns, axis=1)

            # Drop the combined columns
            df.drop(columns=cols_to_combine, inplace=True)

            # Save the processed dataframe back to the CSV
            df.to_csv(file_path, index=False)

            print(f"Processed {file}")
        except pd.errors.ParserError:
            print(f"Couldn't read {file} as a CSV.")
```

```python
import os
import zipfile
from google.colab import files
```

Specify the directory containing the processed files

```python
directory = '/content/'
```

Specify the name of the zip file you want to create

```python
zip_filename = "Arizona_processed_files.zip"
```

Create a zip file

```python
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    for foldername, subfolders, filenames in os.walk(directory):
        for filename in filenames:
            # Ensure you're not trying to zip the zip file itself (if it already exists)
            if filename != zip_filename:
                zipf.write(os.path.join(foldername, filename), os.path.relpath(os.path.join(foldername, filename), directory))
```

Download the zip file

```python
files.download(zip_filename)
```
